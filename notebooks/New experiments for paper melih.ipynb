# Base model feature importances

```python
def plot_by_mismatch(data, pred, data_key='readFraction'):
    num_mism = data['Annotation'].apply(lambda x: len([x[i] for i in range(len(x)) if len(x[i])>2])).values
    mismatches = np.unique(num_mism)
    mm_results = {}
    mismatch_counts = []
    positives_count = []
    pvs = []

    for k in pred.keys():
        mm_results[k] = []

    x = np.logspace(np.log10(10), np.log10(100), 2)
    for mm in mismatches:
        sel = (num_mism == mm)
        mismatch_counts.append(sel.sum())
        positives_count.append((data[data_key][sel] > 0.0).sum())
        tmp = {}

        pv_k = {}
        for k in pred.keys():
            tmp[k] = pred[k].flatten()[sel]

            if k != 'linear-raw-stacker' and k != 'product':
                t2, pv, corr0, corr1,_ = azimuth.util.get_pval_from_predictions(pred['linear-raw-stacker'][sel], 
                                                           pred[k][sel],  
                                                           data[data_key][sel], 
                                                           twotailed=True, method='steiger')
                pv_k[k] = pv
        
        pvs.append(pv_k)  
        roc_results = pp.plot_spearman_with_different_weights(tmp, data[data_key].values.flatten()[sel], 
                                                          weights=x)

        for m in ['linear-raw-stacker', 'CFD', 'HsuZhang', 'CCTOP']:
            mm_results[m].append((((np.array(roc_results[m]) -np.array(roc_results['CCTOP']))/np.array(roc_results['CCTOP']))*100.)[-1])
    
    
    return mm_results, mismatch_counts, positives_count, pvs, mismatches
```
